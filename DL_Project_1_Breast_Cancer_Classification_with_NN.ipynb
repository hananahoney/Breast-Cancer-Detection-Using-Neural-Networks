```python
import sklearn.datasets
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
dataset = sklearn.datasets.load_breast_cancer()
dataset
```

```python
X = pd.DataFrame(data=dataset.data, columns=dataset.feature_names)
```

```python
X.tail()
```

```python
lst = [dataset.target_names[1]]
Y = pd.DataFrame(data=dataset.target, columns=lst)
```

```python
df = X
```

```python
df['benign'] = Y
```

```python
X = df.iloc[:, 0:-1]
y = df.iloc[:, -1]
```

```python
X.head()
```

```python
Y.head()
```

```python
# using the train test split function
X_train, X_test, y_train, y_test = train_test_split(X,y ,
                                   random_state=104,
                                   test_size=0.25,
                                   shuffle=True)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
obj = StandardScaler()

X_train_std = obj.fit_transform(X_train)
X_test_std = obj.fit_transform(X_test)
```

```python
import tensorflow as tf
from tensorflow import keras
```

```python
model = keras.Sequential([
    keras.layers.Flatten(input_shape = (30,)),
    keras.layers.Dense(512, activation='relu'),
    keras.layers.Dense(512, activation='relu'),
    keras.layers.Dense(2, activation='sigmoid'),
])
```

```python
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
model.fit(X_train_std, y_train, epochs=10, validation_split=0.1)
```

```python
loss, accuracy = model.evaluate(X_test_std, y_test)
print(accuracy)
```

```python
y_pred = model.predict(X_test_std)
```
